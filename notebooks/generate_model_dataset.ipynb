```python
import pandas as pd

meta = pd.read_csv("../data/human/IBD_metadata.csv")
meta = meta.set_index("Participant_ID")
meta.head()
```

```python
bacteria_abundance = pd.read_csv("../data/bacteria/IBD_bacteria_abundance_rna_ID.csv")
bacteria_abundance = bacteria_abundance.set_index("Participant_ID")
bacteria_abundance.head()
```

```python
gene_expression = pd.read_csv("../data/human/IBD_gene_expression_selection1.csv")
gene_expression = gene_expression.set_index("Participant_ID")
gene_expression.head()
```

```python
model_dataset = meta.join(bacteria_abundance).join(gene_expression)
model_dataset.to_csv("../data/model/IBD_model_dataset_rna_ID.csv")
```
