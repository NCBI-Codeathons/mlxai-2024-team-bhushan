```python
import pandas as pd

expression = pd.read_csv("../data/human/Priya_table_S14.csv")
expression = expression.set_index("Unnamed: 0")
expression.index.name = "human_gene"
expression = expression.transpose()
expression.index.name = "Participant_ID"
expression.head()
```

```python
selection = 1

interesting = pd.read_csv("../data/human/IBD_interesting_genes.csv")
genes = list(interesting[interesting["Include (1,2)"] <= selection].Name)
genes
```

```python
expression[genes]
```

```python
# store human expression data for selected genes

expression[genes].to_csv("../data/human/IBD_gene_expression_selection1.csv")
```
